## Pandas  groupby
groupby in pandas is a function that lets you group data in a DataFrame based on specific criteria, and then apply aggregate functions to each group. It's a powerful tool for data analysis that allows you to quickly and easily calculate summary statistics for your data.


```python
import numpy as np
import pandas as pd
```

#### Groupby (Applies on Categorical data)

```python
movies = pd.read_csv(" imdb-top-1000.csv")
movies.head(3)
```

```python
movies.groupby('Genre')
```

```python
generes =movies.groupby('Genre')
```

```python
# Applying builtin aggregation fuctions on groupby objects
generes.sum().head(2)
```

```python
#Second Approach
movies.groupby('Genre')['Gross'].sum().sort_values()
```

```python
# find the genre with highest avg IMDB rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# find director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# find the highest rated movie of each genre
movies.head(1)
```

```python
# find number of movies done by each actor
#movies['Star1'].value_counts() # first method
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

### aggregate method

```python
 
# passing dict
movies.groupby('Genre').agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
        
    }

)
```

```python
#Passsing List

movies.groupby('Genre').agg(['min','max','mean','sum','median'])
```

```python
# Adding both the syntax
movies.groupby('Genre').agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# looping on groups
for group , data in movies.groupby('Genre'):
    print(data)
```

```python
# find the highest rated movie of each genre

df = pd.DataFrame(columns=movies.columns)

for group , data in movies.groupby('Genre'):
    df= df.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])
df
```

### split (apply) combine
 


```python
# apply -> builtin function
movies.groupby('Genre').apply(min)
```

```python
# find number of movies starting with A for each group
def foo(group):
    print(group) # type = Dataframe
    return group
```

```python
movies.groupby('Genre').apply(foo)
```

```python
#Custom Logic
def foo(group):
     return group['Series_Title'].str.startswith('A').sum()
    
```

```python
movies.groupby('Genre').apply(foo)
```

```python
# find ranking of each movie in the group according to IMDB score
def rank_movie(group):
    group['genre_rank']=group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
movies.groupby('Genre').apply(rank_movie)
```

```python
 # find normalized IMDB rating group wise
#x normalized = (x – x minimum) / (x maximum – x minimum)
def normal(group):
    group['normal_rating'] = (group['IMDB_Rating']- group['IMDB_Rating'].min())/ (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group

movies.groupby('Genre').apply(normal)
```

### groupby on multiple cols


```python
combo = movies.groupby(['Director','Star1'])
#size
combo.size()
```

```python
#if we want particular combo---> get_group
combo.get_group(('Aamir Khan','Amole Gupte'))
```

```python
# find the most earning actor-> director combo
combo['Gross'].sum().sort_values(ascending=False).head(2)
```

```python
# find the best(in-terms of metascore(avg)) actor->genre combo
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(1)
```

```python
# agg on multiple groupby
combo.agg(['min','max','mean'])
```

# Excercise

```python
ipl = pd.read_csv("deliveries.csv")
ipl.head(2)
```

```python
ipl.shape # ball by ball dataset
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).reset_index().head(10)
```

```python
# find the batsman with max no of sixes
six = ipl[ipl['batsman_runs']==6]

six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(2)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs

temp = ipl[ipl['over']>15]
temp = temp[(temp['batsman_runs'] == 4) | (temp['batsman_runs'] == 6)]

temp.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# find V Kohli's record against all teams

temp = ipl[ipl['batsman'] == 'V Kohli']

temp.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).reset_index().head(2)
```

```python
# Create a function that can return the highest score of any batsman

temp = ipl[ipl['batsman'] == 'V Kohli']

temp.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
    temp = ipl[ipl['batsman'] == batsman]
    return  temp.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest('DA Warner')
```
